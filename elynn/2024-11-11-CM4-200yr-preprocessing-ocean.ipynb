```python
CONFIG_YAML = "/home/jovyan/full-model/scripts/data_process/configs/CM4-pre-industrial-control-ocean-sample.yaml"
config = OceanDatasetConfig.from_file(CONFIG_YAML)

# this is the key in config.runs
RUN_KEY = "2024-11-11-CM4-ocean-200yr-preprocessed"
run_directory = config.runs[RUN_KEY]

output_store = os.path.join(config.data_output_directory, f"{RUN_KEY}.zarr")

print(f"raw data dir is {run_directory}")
print(f"output store is {output_store}")

data_config = config.dataset_computation
dask_config = config.dask
fs = fsspec.filesystem(
    "s3",
    profile="ai2",  ## This is the profile name you configured above.
    client_kwargs={"endpoint_url": "https://nyu1.osn.mghpcc.org"} # This is the endpoint for the m2lines osn pod
)

print(f"dask config is {dask_config}")
```

```python
backend_kwargs={
        "storage_options": {
            "key": "KEY",
            "secret": "SECRET",
            "client_kwargs": {"endpoint_url": "https://nyu1.osn.mghpcc.org"}
        }
}
```

```python
if dask_config is not None:
    client = dask_config.get_client()
else:
    client = None

client
```

```python
ds = compute_lazy_dataset(
    config=data_config, 
    run_directory=run_directory,
    fs=fs,
    backend_kwargs=backend_kwargs,
)
ds
```

```python
n_partitions = config.n_split

# set RESUME_PARTITION to the last successfully completed segment number 
# use this in case computation hangs and you have to restart the kernel
RESUME_PARTITION = 0

for i in range(RESUME_PARTITION, n_partitions):
    if i == 0:
        ds.partition.initialize_store(output_store)

    print(f"Writing segment {i + 1} / {n_partitions}")
    ds.partition.write(
        output_store,
        n_partitions,
        ["time"],
        i,
        collect_variable_writes=True,
    )
```

```python
# close the cluster
dask_config.close_cluster()
```
